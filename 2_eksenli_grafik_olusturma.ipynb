```python
import pandas as pd
import matplotlib.pyplot as plt

# Verileri giriniz
kazanc_durumu = ['0', '-1', '5', '-8', '7']
aylar = ['Temmuz', 'Agustos', 'Eylül', 'Ekim', 'Aralık']

# Sınıflandırma yapmak için bir fonksiyon
def kazanc_siniflandir(deger):
    deger = int(deger)  # String'i integer'a dönüştür
    if deger < 0:
        return 'ZARAR'
    elif deger == 0:
        return 'NÖTR'
    elif deger > 0:
        return 'KAR'

# Kazanc durumuna göre sınıflandırma uygulama
kazanc_siniflar = [kazanc_siniflandir(deger) for deger in kazanc_durumu]

# Verileri ve sınıfları içeren bir veri çerçevesi oluşturma
df_kazanc = pd.DataFrame({'Zaman': aylar, 'Kar/Zarar': kazanc_durumu, 'Finans': kazanc_siniflar})

# Renk tanımları
renk = {'KAR': 'green', 'ZARAR': 'red', 'NÖTR': 'gray'}

# Görsel olarak tabloyu çizdirme
fig, ax = plt.subplots()
bars = plt.bar(aylar, [int(deger) for deger in kazanc_durumu], color=[renk[sirket_mali_durumu] for sirket_mali_durumu in kazanc_siniflar])
plt.xlabel('Zaman')
plt.ylabel('Mali Durum')
plt.title('Şirket Bilançosu')

# Etiketleri ekleme
for bar, sirket_mali_durum in zip(bars, kazanc_siniflar):
    ax.text(bar.get_x() + bar.get_width() / 2, bar.get_height(), sirket_mali_durum, ha='center', va='bottom', color='black', fontsize=8)

plt.show()
```
